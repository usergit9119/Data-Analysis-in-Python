<h2>Линейная регрессия</h2>

<h3>Что такое Регрессия?</h3>

<ul>
    <li><p>Регрессия ищет зависимость между переменными </p></li>
    <li><p>регрессия помогает нам построить математическую модель, которая объединяет различные наблюдения в единое целое, что позволяет нам лучше понять взаимосвязи между переменными и использовать эти знания для прогнозирования.</p></li>
    <li><p>Регрессия рассматривает некоторые наблюдения и ряд явлений. Каждое наблюдение имеет две и более переменных</p></li>
    <li><p>Другими словами нужно найти функцию, которая отображает зависимость от одних переменных от других</p></li>
    <li><p><b>Зависимые данные</b> называются зависимыми переменными, выходами или ответами.
    <b>Независимые данные</b> называются независимыми переменными, входами или предсказателями.</p></li>
    
</ul>

<h3>Когда нужна Регрессия?</h3>

<p>Регрессия полезна для прогнозирования ответа на новые условия. Можно угадать потребление электроэнергии в жилом доме из данных температуры, времени суток и количества жильцов.</p>


<h3>Постанковка проблемы</h3>

<p>Линейная регрессия представляет функцию y набор независимых переменных $x = (x_{1}, _{...}, x_{r})$, где r - число предсказателей, предполагает, что линейное отношение между y и x. Это и есть <b>Уравнение регрессии</b>: $$ y = \beta_0 + \beta_1 x_1 + \dots + \beta_r x_r + \varepsilon $$
<ul>
Где:
    <li>$Y$ -зависимая переменная (целевая переменная)</li>
    <li>$X_{1}, X_{2}, ..., X_{n}$ - независимые переменные (предикаторы)</li>
    <li>$\beta_0, \beta_1, ... ,\beta_n$ - коэффиценты регрессии,которые определяют влияние каждой независимой переменной на зависимую переменную. </li>
    <li>$\varepsilon$ - случайная ошибка, которая представляет расхождение между фактическими и прогнозируемыми значениями зависимой перменной</li>
</ul>
</p>

<h4>Оценочные функции</h4>

<p>Оценочная функция в контексте линейной регрессии используется для прогнозирования значений зависимой переменной на основе значений независимых переменных. В основе её использования лежат коэффициенты регрессии, которые были оценены в процессе обучения модели на обучающей выборке данных.<br>
    Для прогнозирования значений зависимой переменной 
$Y$ для нового наблюдения $x$, мы подставляем соответствующие значения $x$ в оценочную функцию $f(x)$. Это даёт нам ожидаемое значение 
$Y$ на основе данных $x$ и параметров модели: $$Y_{predicted} = f(x) = b_{0} + b_{1}x_{1} + ... + b_{r}x_{r}$$
Где:
<ul> 
    <li>$Y_{predicted}$ - прогнозируемое значение</li>
    <li>$x_{1},x_{2},...,x_{r}$ - значения независимых переменных для нового наблюдения.</li>
    <li>$b_{0},b_{1},...,b_{r}$ - коэффиценты регрессии, оцененные в процессе обучения</li>
    
</ul>
Таким образом, оценочная функция позволяет нам использовать модель линейной регрессии для прогнозирования значений зависимой переменной на основе новых данных, которые не были использованы при обучении модели. Это делает оценочную функцию полезным инструментом в анализе и прогнозировании данных.
</p>

<h4>Остаток регрессии</h4>

Для каждого результата наблюдения i = 1, ..., n, <b>Оценочный(предсказанный ответ)</b> $f(x_{i})$ ,должен быть как можно ближе к соответствующему <b>фактическому ответ</b> $y_{i}$. Разницы $y_{i} - f(x_{i})$ для всех результатов наблюдений называются <b>остатками</b>. Регрессия определяет лучшие <b>прогнозируемые весы измерения</b>, которые соответствуют наименьшими остаткам.

в общем случае, чем меньше остаток регрессии, тем более точно модель линейной регрессии соответствует данным, и, следовательно, результат наблюдения может считаться более корректным. Остатки регрессии представляют собой разницу между фактическими значениями зависимой переменной и значениями, предсказанными моделью. Если остатки маленькие, это означает, что модель хорошо предсказывает или объясняет данные.

Однако следует учитывать, что оценка корректности результатов наблюдения и модели зависит от контекста и целей исследования. Например, если остатки регрессии очень маленькие, это может указывать на переобучение модели, когда она слишком точно подстраивается под обучающие данные и неспособна обобщать на новые данные. В таких случаях модель может быть некорректной.

<h2>SSE,SSR,SST и связанные метрики.</h2>

<b><h3>SSE (Сумма квадратов ошибок отклонений):</h3> SSE - является важной метрикой, используемой в статистике и машинном обучении для оценки качества модели. Вот полное описание SSE:</b>
<ol>
    <li><b>Определение:</b> SSE измеряет сумму квадратов разностей между фактическими значениями зависимой переменной и предсказанными значениями, полученными с использованием модели. Формально это выражается как: $$ SSE = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$$</li>
    <ul>
        <li>Знак $\sum$ обозначает сумму.</li>
        <li>$n$ - количество наблюдений</li>
        <li>$y_{i}$ - фактическое значение зависимой переменной для $i$-го наблюдения</li>
        <li>$\hat{y}_i$ - предсказанное значение зависимой переменной для $i$-го наблюдения</li>
        <li>а именно возводя разности в квадрат, мы учитываем их абсолютные величины и их направления, чтобы избежать компенсации положительных и отрицательных отклонений друг от друга. Этот шаг и минимизирует сумму квадратов ошибок отклонений  </li>
    </ul>
    <li><b>Интерпретация:</b> SSE показывает, насколько хорошо модель предсказывает фактические данные. Чем меньше значение SSE, тем ближе предсказанные значения к фактическим, и тем лучше модель описывает данные.</li>
    <li><b>Цель оптимизации:</b> В контексте метода наименьших квадратов (МНК), который широко применяется в линейной регрессии, целью является минимизация SSE. Подбор коэффициентов модели осуществляется таким образом, чтобы минимизировать SSE и, следовательно, улучшить качество модели.</li>
    <li><b>Необъясненная изменчивость:</b> SSE измеряет необъясненную моделью изменчивость зависимой переменной, то есть это и есть <i>разница между фактическими значениями и предсказанными</i>. Эта разница (или остаток) представляет собой необъясненную изменчивость.По сути, необъясненная изменчивость указывает на ту часть вариации зависимой переменной, которую модель не может объяснить. Если модель хорошо описывает данные, то необъясненная изменчивость будет относительно небольшой. Если модель плохо описывает данные, то необъясненная изменчивость будет значительной. Необъясненная изменчивость также известна как сумма квадратов ошибок (SSE). <br>
    Таким образом, необъясненная изменчивость — это общее понятие, которое вконтексте задач включает в себя погрешности приборов измерений, отсутствие важной информации или переменных в модели, ошибки в спецификации модели, а также случайные колебания и шум в данных. Минимизация этой необъясненной изменчивости помогает улучшить точность и качество модели.</li>
    <li><b>Связь с другими метриками:</b>SSE тесно связан с другими метриками, такими как сумма квадратов объясненных отклонений (SSR) и общая сумма квадратов (SST). Вся вариация в данных (SST) может быть разделена на объясненную моделью вариацию (SSR) и необъясненную вариацию (SSE): $$SST = SSR + SSE$$</li>
    <li><b>Ограничения SSE</b>: Поскольку SSE представляет собой сумму квадратов разностей, величина этого значения напрямую зависит от масштаба данных. Если значения переменных имеют разные масштабы, SSE будет искажаться, и его невозможно будет использовать для прямого сравнения разных моделей. Из-за своей зависимости от масштаба, SSE не всегда подходит для сравнения разных моделей на разных наборах данных. Это может привести к ложным выводам о качестве модели. Например: Данные могут иметь разные диапазоны значений, например, от 0 до 1 или от 0 до 1000. Если данные в одном наборе данных имеют широкий диапазон значений, а в другом — узкий, это может повлиять на оценку модели. <br>
    хотя минимизация SSE может помочь в достижении высокой точности, это не всегда означает, что модель имеет хорошее общее качество или способность к обобщению на другие данные.  Например, если модель слишком сильно адаптирована к данным (перенастройка или переобучение), это может привести к тому, что модель будет хорошо предсказывать конкретные данные, но она может плохо работать с новыми, ранее невиданными данными, поэтому хорошие практики моделирования должны включать в себя баланс между минимизацией SSE и поддержанием высокой обобщающей способности модели(способность предсказывать новые, ранее невиданные данные).</li>
</ol>
<b> <h3>Сумма квадратов регрессии (SSR):</h3> SSR — это показатель того, насколько хорошо модель регрессии объясняет данные, и помогает оценить эффективность модели.</b><br>
Вот основные моменты, связанные с SSR:
<ol>
    <li><b>Определение:</b>Сумма квадратов регрессии (SSR) или "объясненная сумма квадратов", представляет собой характеристику, которая измеряет долю изменчивости зависимой переменной, которая объясняется моделью регрессии. Другими словами, SSR показывает, насколько хорошо модель объясняет фактические данные.Формула для расчета SSR: $$ SSR = \sum_{i=1}^{n} (\hat{y}_i - \overline{y})^2$$</li>
    <ul>
        <li>Знак $\sum$ обозначает сумму.</li>
        <li>$N$ - количество наблюдений</li>
        <li>$\overline{y}$ - среднее значение зависимой переменной по всем наблюдениям.</li>
        <li>$\hat{y}_i$ - предсказанное значение зависимой переменной для $i$-го наблюдения</li>
        <li>Возведение разностей в квадрат позволяет избежать иллюзии точной модели. Без возведения в квадрат положительные и отрицательные разности могли бы компенсировать друг друга, создавая иллюзию точной модели. Это происходит потому, что при наличии больших положительных и отрицательных отклонений их сумма может приблизиться к нулю, давая ложное впечатление, что предсказания модели близки к фактическим значениям.</li>
    </ul>
    <li><b>Связь с другими метриками:</b>SSR можно рассматривать как разницу между полной суммой квадратов (SST) и необъясненной суммой квадратов (SSE):$$SSR = SST - SSE$$</li>
    <li><b>Объясненная изменчивость:</b> SSR показывает, какую часть изменчивости зависимой переменной можно объяснить моделью регрессии. Чем выше значение SSR, тем больше доля объясненной изменчивости, что указывает на более качественную модель. Когда мы говорим о том, что сумма квадратов регрессии (SSR) измеряет долю изменчивости зависимой переменной, мы имеем в виду, что SSR показывает, какую часть общей изменчивости в зависимой переменной (то есть, насколько значения зависят от друг от друга или от среднего значения) можно объяснить моделью регрессии.

В модели регрессии у нас есть зависимая переменная (например, результат измерений) и независимые переменные (например, объясняющие факторы). Модель пытается предсказать зависимую переменную на основе независимых переменных.

Когда модель хорошо подогнана, она сможет объяснить большую часть изменчивости зависимой переменной. Это значит, что изменения в зависимой переменной хорошо коррелируют с изменениями в независимых переменных, которые используются в модели. SSR показывает, насколько успешно модель объясняет зависимую переменную.</li>
    <li><b>Ограничения SSR:</b>
    <ul>
    <li><b>Чувствительность к шуму в данных:</b> В случаях, когда в данных много шума (случайных колебаний), высокий SSR не всегда указывает на хорошую модель. Иногда модель может просто подогнать шум в данных, что приведет к переобучению.</li>
    <li><b>Ограниченная интерпретация:</b> Хотя высокий SSR указывает на объясненную долю изменчивости, он не предоставляет информации о предсказательной силе модели на новых данных.</li>
    <li><b>Сложность модели:</b> Чем сложнее модель(много параметров,использование сложных функций/структур), тем больше она склонна к подстраиванию под шумы(искажения в данных) в наборе данных. Это явление называется переобучением (overfitting). Когда модель усложняется, чтобы лучше объяснить наблюдаемые данные, это часто приводит к увеличению значения суммы объясненных квадратов (SSR). Однако это не всегда означает, что модель действительно улучшилась в качественном плане. На самом деле из-за cложной модели, которая идеально подходит к обучающим данным, может подогнаться к случайным колебаниям и шуму в данных, а не только к реальным тенденциям. В результате такая модель будет иметь высокое значение SSR, но низкую обобщающую способность — она будет плохо предсказывать новые данные. Таким образом, при выборе модели и оптимизации ее параметров важно находить баланс между объясненной и необъясненной изменчивостью, чтобы избежать чрезмерного подстраивания модели под обучающий набор данных. Это поможет добиться более высокой обобщающей способности и обеспечить более устойчивые и надежные предсказания на новых, невиданных данных. Также В случае очень сложных моделей становится труднее интерпретировать, какие факторы действительно влияют на предсказания модели.</li>
    <li>Введение ложных связей и связь корреляции: Сложные модели могут вводить ложные связи между переменными, особенно если данные имеют шум или выбросы. Ложные связи возникают, когда модель выявляет корреляции между переменными, которые на самом деле не существуют в реальности. Эти корреляции могут быть результатом случайных колебаний в данных или артефактов выборки. Такая ситуация может привести к неверной интерпретации модели и ее результатов. Например, модель может предположить, что есть связь между двумя переменными, которая в действительности отсутствует. Это может повлиять на предсказательную способность модели, особенно на новых, ранее невиданных данных.</li>
    <ul>
    </li>
    <li><b>Связь с коэффиицентом детерминации:</b> Для оценки этой способности регрессии объяснять зависимую переменную, используется коэффициент детерминации $R^2$, который равен отношению суммы квадратов регрессии (SSR) к полной сумме квадратов (SST): $$R^2 = \frac{SSR}{SST}$$ Коэффициент детерминации — это мера, которая определяет, насколько хорошо модель объясняет вариацию в зависимой переменной. Это показатель качества подгонки модели к данным.
Высокое значение $R^2$ указывает на то, что модель объясняет значительную часть изменчивости зависимой переменной, и, следовательно, модель хорошо подогнана к данным.</li>
</ol>
Характеристики коэффициента детерминации:
<ul>
    <li><b>Диапазон значений:</b>R^2 варьируется от 0 до 1. Чем выше значение R^2, тем лучше модель объясняет вариацию в данных. Если R^2 равно 1, это означает, что модель идеально описывает данные.</li>
    <li><b>Формула:</b> R^2 вычисляется по формуле: $$R^2 = 1 - \frac{SSE}{SST}$$</li>
    <li><b>Интерпритация сложных моделей:</b>При использовании сложных моделей стоит осторожно интерпретировать значение $R^2$, т.к. оно может быть искажено из-за переобучения модели. Вследствии этого, модель не сможет обобщать хорошо новые данные. На новых данных модель может показать плохие результаты из-за того, что она слишком точно подстроилась под шумы или особенности конкретных данных обучающего набора.</li>
    <li><b>Применимость:</b>R-квадрат помогает оценить, насколько хорошо модель соответствует данным. Чем ближе значение R-квадрата к 1, тем лучше модель объясняет вариацию зависимой переменной. Путем сравнения значений R-квадрата разных моделей можно определить, какая из них лучше соответствует данным. Модель с более высоким значением R-квадрата обычно считается более предпочтительной.</li>
</ul>

<b><h3>SSE (Сумма квадратов ошибок отклонений):</h3> SSE - является важной метрикой, используемой в статистике и машинном обучении для оценки качества модели. Вот полное описание SSE:</b>
<ol>
    <li><b>Определение:</b> SSE измеряет сумму квадратов разностей между фактическими значениями зависимой переменной и предсказанными значениями, полученными с использованием модели. Формально это выражается как: $$ SSE = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$$</li>
    <ul>
        <li>Знак $\sum$ обозначает сумму.</li>
        <li>$n$ - количество наблюдений</li>
        <li>$y_{i}$ - фактическое значение зависимой переменной для $i$-го наблюдения</li>
        <li>$\hat{y}_i$ - предсказанное значение зависимой переменной для $i$-го наблюдения</li>
        <li>а именно возводя разности в квадрат, мы учитываем их абсолютные величины и их направления, чтобы избежать компенсации положительных и отрицательных отклонений друг от друга. Этот шаг и минимизирует сумму квадратов ошибок отклонений  </li>
    </ul>
    <li><b>Интерпретация:</b> SSE показывает, насколько хорошо модель предсказывает фактические данные. Чем меньше значение SSE, тем ближе предсказанные значения к фактическим, и тем лучше модель описывает данные.</li>
    <li><b>Цель оптимизации:</b> В контексте метода наименьших квадратов (МНК), который широко применяется в линейной регрессии, целью является минимизация SSE. Подбор коэффициентов модели осуществляется таким образом, чтобы минимизировать SSE и, следовательно, улучшить качество модели.</li>
    <li><b>Необъясненная изменчивость:</b> SSE измеряет необъясненную моделью изменчивость зависимой переменной, то есть это и есть <i>разница между фактическими значениями и предсказанными</i>. Эта разница (или остаток) представляет собой необъясненную изменчивость.По сути, необъясненная изменчивость указывает на ту часть вариации зависимой переменной, которую модель не может объяснить. Если модель хорошо описывает данные, то необъясненная изменчивость будет относительно небольшой. Если модель плохо описывает данные, то необъясненная изменчивость будет значительной. Необъясненная изменчивость также известна как сумма квадратов ошибок (SSE). <br>
    Таким образом, необъясненная изменчивость — это общее понятие, которое вконтексте задач включает в себя погрешности приборов измерений, отсутствие важной информации или переменных в модели, ошибки в спецификации модели, а также случайные колебания и шум в данных. Минимизация этой необъясненной изменчивости помогает улучшить точность и качество модели.</li>
    <li><b>Связь с другими метриками:</b>SSE тесно связан с другими метриками, такими как сумма квадратов объясненных отклонений (SSR) и общая сумма квадратов (SST). Вся вариация в данных (SST) может быть разделена на объясненную моделью вариацию (SSR) и необъясненную вариацию (SSE): $$SST = SSR + SSE$$</li>
    <li><b>Ограничения SSE</b>: Поскольку SSE представляет собой сумму квадратов разностей, величина этого значения напрямую зависит от масштаба данных. Если значения переменных имеют разные масштабы, SSE будет искажаться, и его невозможно будет использовать для прямого сравнения разных моделей. Из-за своей зависимости от масштаба, SSE не всегда подходит для сравнения разных моделей на разных наборах данных. Это может привести к ложным выводам о качестве модели. Например: Данные могут иметь разные диапазоны значений, например, от 0 до 1 или от 0 до 1000. Если данные в одном наборе данных имеют широкий диапазон значений, а в другом — узкий, это может повлиять на оценку модели. <br>
    хотя минимизация SSE может помочь в достижении высокой точности, это не всегда означает, что модель имеет хорошее общее качество или способность к обобщению на другие данные.  Например, если модель слишком сильно адаптирована к данным (перенастройка или переобучение), это может привести к тому, что модель будет хорошо предсказывать конкретные данные, но она может плохо работать с новыми, ранее невиданными данными, поэтому хорошие практики моделирования должны включать в себя баланс между минимизацией SSE и поддержанием высокой обобщающей способности модели(способность предсказывать новые, ранее невиданные данные).</li>
</ol>
<b> <h3>Сумма квадратов регрессии (SSR):</h3> SSR — это показатель того, насколько хорошо модель регрессии объясняет данные, и помогает оценить эффективность модели.</b><br>
Вот основные моменты, связанные с SSR:
<ol>
    <li><b>Определение:</b>Сумма квадратов регрессии (SSR) или "объясненная сумма квадратов", представляет собой характеристику, которая измеряет долю изменчивости зависимой переменной, которая объясняется моделью регрессии. Другими словами, SSR показывает, насколько хорошо модель объясняет фактические данные.Формула для расчета SSR: $$ SSR = \sum_{i=1}^{n} (\hat{y}_i - \overline{y})^2$$</li>
    <ul>
        <li>Знак $\sum$ обозначает сумму.</li>
        <li>$N$ - количество наблюдений</li>
        <li>$\overline{y}$ - среднее значение зависимой переменной по всем наблюдениям.</li>
        <li>$\hat{y}_i$ - предсказанное значение зависимой переменной для $i$-го наблюдения</li>
        <li>Возведение разностей в квадрат позволяет избежать иллюзии точной модели. Без возведения в квадрат положительные и отрицательные разности могли бы компенсировать друг друга, создавая иллюзию точной модели. Это происходит потому, что при наличии больших положительных и отрицательных отклонений их сумма может приблизиться к нулю, давая ложное впечатление, что предсказания модели близки к фактическим значениям.</li>
    </ul>
    <li><b>Связь с другими метриками:</b>SSR можно рассматривать как разницу между полной суммой квадратов (SST) и необъясненной суммой квадратов (SSE):$$SSR = SST - SSE$$</li>
    <li><b>Объясненная изменчивость:</b> SSR показывает, какую часть изменчивости зависимой переменной можно объяснить моделью регрессии. Чем выше значение SSR, тем больше доля объясненной изменчивости, что указывает на более качественную модель. Когда мы говорим о том, что сумма квадратов регрессии (SSR) измеряет долю изменчивости зависимой переменной, мы имеем в виду, что SSR показывает, какую часть общей изменчивости в зависимой переменной (то есть, насколько значения зависят от друг от друга или от среднего значения) можно объяснить моделью регрессии.

В модели регрессии у нас есть зависимая переменная (например, результат измерений) и независимые переменные (например, объясняющие факторы). Модель пытается предсказать зависимую переменную на основе независимых переменных.

Когда модель хорошо подогнана, она сможет объяснить большую часть изменчивости зависимой переменной. Это значит, что изменения в зависимой переменной хорошо коррелируют с изменениями в независимых переменных, которые используются в модели. SSR показывает, насколько успешно модель объясняет зависимую переменную.</li>
    <li><b>Ограничения SSR:</b>
    <ul>
    <li><b>Чувствительность к шуму в данных:</b> В случаях, когда в данных много шума (случайных колебаний), высокий SSR не всегда указывает на хорошую модель. Иногда модель может просто подогнать шум в данных, что приведет к переобучению.</li>
    <li><b>Ограниченная интерпретация:</b> Хотя высокий SSR указывает на объясненную долю изменчивости, он не предоставляет информации о предсказательной силе модели на новых данных.</li>
    <li><b>Сложность модели:</b> Чем сложнее модель(много параметров,использование сложных функций/структур), тем больше она склонна к подстраиванию под шумы(искажения в данных) в наборе данных. Это явление называется переобучением (overfitting). Когда модель усложняется, чтобы лучше объяснить наблюдаемые данные, это часто приводит к увеличению значения суммы объясненных квадратов (SSR). Однако это не всегда означает, что модель действительно улучшилась в качественном плане. На самом деле из-за cложной модели, которая идеально подходит к обучающим данным, может подогнаться к случайным колебаниям и шуму в данных, а не только к реальным тенденциям. В результате такая модель будет иметь высокое значение SSR, но низкую обобщающую способность — она будет плохо предсказывать новые данные. Таким образом, при выборе модели и оптимизации ее параметров важно находить баланс между объясненной и необъясненной изменчивостью, чтобы избежать чрезмерного подстраивания модели под обучающий набор данных. Это поможет добиться более высокой обобщающей способности и обеспечить более устойчивые и надежные предсказания на новых, невиданных данных. Также В случае очень сложных моделей становится труднее интерпретировать, какие факторы действительно влияют на предсказания модели.</li>
    <li><b>Введение ложных связей и связь корреляции:</b>Корреляция — это статистическая мера, которая описывает силу и направление отношения между двумя переменными. В контексте анализа данных, корреляция измеряет, как изменение одной переменной связано с изменением другой переменной.

Существует несколько типов корреляции:
<ul>
    <li><b>Положительная корреляция:</b> когда одна переменная увеличивается, другая также увеличивается. Например, вес человека и его рост.</li>
    <li><b>Отрицательная корреляция:</b> когда одна переменная увеличивается, другая уменьшается. Например, количество часов сна и уровень стресса.</li>
    <li><b>Нулевая корреляция:</b> когда нет никакой связи между переменными. Изменение одной переменной не связано с изменением другой.</li>
</ul>
Величина корреляции обычно выражается коэффициентом корреляции, который может варьироваться от -1 до 1:
<ul>
    <li>1: указывает на совершенную положительную корреляцию.</li>
    <li>-1: указывает на совершенную отрицательную корреляцию.</li>
    <li>0: указывает на отсутствие корреляции.</li>
</ul>
Высокий коэффициент корреляции не обязательно указывает на причинно-следственную связь между переменными. Вместо этого он показывает только, что изменения в одной переменной связаны с изменениями в другой переменной. в контексте SSR (сумма объясненных квадратов), чем выше положительная корреляция между предсказанными значениями модели и фактическими значениями зависимой переменной, тем больше модель объясняет общую изменчивость, и тем выше SSR.<br>
Ложные связи (или спурные связи) возникают, когда две переменные кажутся связанными или коррелированными, но их связь является результатом случайности или третьего фактора, а не прямой причинно-следственной зависимости между ними. Это может быть проблематично при построении статистических моделей, так как может привести к неправильным выводам о влиянии одной переменной на другую.</li>
    Вот несколько ситуаций, когда могут возникать ложные связи:
<ul>
    <li><b>Общая причина:</b> Две переменные могут быть связаны из-за того, что на них влияет третья переменная. Например, корреляция между продажами мороженого и количеством утоплений может быть обусловлена третьей переменной — температурой воздуха. В жаркую погоду больше людей покупают мороженое и больше плавают, что может привести к увеличению числа утоплений.</li>
    <li><b>Случайные корреляции:</b> В больших наборах данных может произойти случайное совпадение значений двух переменных, создавая иллюзию связи. Это особенно вероятно, если исследуется множество переменных одновременно. Например, представьте исследование, в котором вы изучаете данные о здоровье, доходах и образе жизни группы людей. Если вы анализируете множество переменных, вы можете обнаружить, что, например, количество зубных пломб у людей связано с количеством телевизоров в доме. На первый взгляд, такая корреляция кажется странной или бессмысленной, и вероятно, что это случайное совпадение. Тем не менее, если вы ищете корреляции между множеством переменных, такие случайные связи могут возникнуть просто из-за большого количества сравнений, даже если на самом деле нет никакой истинной взаимосвязи между этими переменными.<br>
Для того чтобы избежать ложных корреляций, важно применять корректировки на множественное тестирование или использовать другие методы статистического анализа, чтобы убедиться, что обнаруженные корреляции являются статистически значимыми и имеют практический смысл.</li>
    <li><b>Ошибки измерений или сбора данных.</b> Ложные связи и корреляции возникают в анализе данных по нескольким причинам. Например, ошибки при измерении данных могут привести к ложным корреляциям, если некоторые измерения оказываются неточными или случайными. Например, если при измерении температуры датчики были неверно откалиброваны, это может привести к ложной корреляции между температурой и другой переменной, когда такой связи на самом деле нет. <br>Также, небольшое количество выбросов(аутлайеров) в данных может существенно повлиять на корреляцию между переменными, создавая ложные связи. 
<b>Аутлайеры или выбросы</b> — это данные, которые резко отклоняются от остальной части набора данных. Они могут существенно повлиять на корреляцию между переменными, создавая ложные связи.
<ul>
    <li>Искажение корреляции: Аутлайеры могут существенно искажать корреляцию между переменными. Например, если у вас есть несколько выбросов в одном конце набора данных, это может изменить наклон линейной регрессии и привести к ложным выводам о связи между переменными.</li>

<li>Сбивка тренда: Выбросы могут внести значительный вклад в общий тренд или модель данных, что может привести к неверным прогнозам или ошибочным выводам о корреляции.</li>

<li>Увеличение дисперсии: Аутлайеры могут существенно увеличить дисперсию данных, что может затруднить обнаружение истинных корреляций или привести к увеличению ложных положительных результатов.</li>
</ul>
<b>Обработка аутлайеров:</b><br>
Обнаружение аутлайеров: Используйте визуализацию данных (например, графики рассеяния, диаграммы размаха) или статистические методы (например, межквартильное расстояние, стандартное отклонение) для обнаружения выбросов.


Проверка достоверности: Убедитесь, что аутлайеры не являются результатом ошибок в данных или сборе данных. Если выбросы вызваны ошибками, их следует исправить.

Удаление или корректировка: Если аутлайеры действительно являются аномалиями и мешают анализу, вы можете рассмотреть возможность их удаления или корректировки. Будьте осторожны, чтобы не удалять данные, которые могут быть важными для анализа.

Масштабирование данных: Масштабирование данных может помочь снизить влияние аутлайеров на анализ, хотя это не всегда решает проблему.

Устойчивые к выбросам методы: Рассмотрите использование методов статистического анализа и машинного обучения, устойчивых к выбросам, таких как методы медианного взвешивания или регрессии с устойчивым влиянием.

Проведение кросс-валидации: Кросс-валидация может помочь проверить, насколько устойчивы результаты анализа к выбросам, распределенным по разным выборкам данных.

<i>Обработка выбросов требует осторожности, поскольку иногда они могут содержать важную информацию о данных или процессе, который вы исследуете. Правильный подход зависит от конкретной задачи и данных, с которыми вы работаете.</i>
</li>
<b>Примеры аутлайеров:</b>
Аутлайеры в измерениях температуры:

В наборе данных о ежедневных температурах в городе может быть аномальное измерение, например, внезапно высокая температура, которая отличается от всех других значений.
Аутлайеры в финансовых данных:

В наборе данных о ценах на акции может быть день с экстремально высоким или низким значением цены, который сильно отличается от других дней.
Аутлайеры в данных о времени поездки:

В наборе данных о времени, затраченном на поездку на работу, может быть одно наблюдение с очень длительным временем, например, из-за затора или аварии.

<b>Визуализация аутлайеров:</b>
Диаграммы размаха (boxplot): Эти графики показывают распределение данных и могут визуально выделить аутлайеры за пределами "усов" диаграммы.

Графики рассеяния: При построении графика рассеяния можно увидеть наблюдения, которые сильно отклоняются от общей тенденции данных.
</ul>
<b>Взаимозависимость признаков:</b> Взаимозависимость признаков, также известная как мультиколлинеарность, происходит, когда независимые переменные в модели имеют высокую корреляцию между собой. Это может привести к нестабильным оценкам коэффициентов и осложнениям при интерпретации результатов анализа данных. В контексте машинного обучения и статистики это может вызвать следующие проблемы:
<ul>
    <li><b>Нестабильность коэффициентов:</b>Если две или более переменных сильно коррелированы, изменение одной из них может привести к значительному изменению оценки коэффициентов других переменных. Это делает модель более чувствительной к шуму в данных и может привести к неправильным выводам.</li>
    <li><b>Трудности интерпретации:</b>Взаимозависимость признаков затрудняет интерпретацию влияния отдельных переменных на зависимую переменную. Это потому, что сложно определить, какая переменная оказывает реальный эффект, а какая влияет только из-за своей взаимозависимости с другой переменной.</li>
    <li><b>Уменьшение качества модели:</b>Модель, в которой присутствует мультиколлинеарность, может иметь более низкое качество прогнозов, так как она становится более чувствительной к шуму и небольшим изменениям в данных.</li>
</ul>
    </li>
    <li><b>Связь с коэффиицентом детерминации:</b> Для оценки этой способности регрессии объяснять зависимую переменную, используется коэффициент детерминации $R^2$, который равен отношению суммы квадратов регрессии (SSR) к полной сумме квадратов (SST): $$R^2 = \frac{SSR}{SST}$$ Коэффициент детерминации — это мера, которая определяет, насколько хорошо модель объясняет вариацию в зависимой переменной. Это показатель качества подгонки модели к данным.
Высокое значение $R^2$ указывает на то, что модель объясняет значительную часть изменчивости зависимой переменной, и, следовательно, модель хорошо подогнана к данным.</li>
</ol>
Характеристики коэффициента детерминации:
<ul>
    <li><b>Диапазон значений:</b>R^2 варьируется от 0 до 1. Чем выше значение R^2, тем лучше модель объясняет вариацию в данных. Если R^2 равно 1, это означает, что модель идеально описывает данные.</li>
    <li><b>Формула:</b> R^2 вычисляется по формуле: $$R^2 = 1 - \frac{SSE}{SST}$$</li>
</ul>

<h2>Метод наименьших квадратов</h2>
Для получения лучших весов, вам нужно минимизировать сумму сумму квадратов ошибок отклоенений (SSE) для всех результатов наблюдений. Этот подход называется <b>методом наименьших квадратов</b>.
Метод наименьших квадратов (МНК) - это основной метод, используемый в линейной регрессии для оценки коэффициентов модели. Его цель состоит в том, чтобы найти такие коэффициенты $\beta_{0},\beta_{1},...,\beta_{r}$, которые минимизируют сумму квадратов остатков, или разностей между фактическими и предсказанными значениями зависимой переменной.

Вот шаги метода наименьших квадратов в линейной регрессии:
<ol>
    <li><b>Формулировка модели:</b>  Вы начинаете с формулировки модели линейной регрессии, которая представляет зависимость между зависимой переменной $y$ и независимыми переменными $x_{1},x_{2},...,x_{r}$. Выглядет это обычно так: $y = \beta_0 + \beta_1 x_1 + \dots + \beta_r x_r + \varepsilon$.</li>
    <li><b>Оценка коэффицентов:</b>Задача состоит в том, чтобы найти такие значения коэффициентов $\beta_{0},\beta_{1},...,\beta_{r}$, которые минимизируют сумму квадратов остатков. Формально, мы минимизируем функцию потерь, используя - <b>сумму квадратов ошибок отклонений</b>: $$E_{1}=\sum_{i=1}^{n} (y_i - \hat{y}_i)^2$$</li>
    <ul>
        <li>Знак $\sum$ обозначает сумму.</li>
        <li>$n$ - количество наблюдений</li>
        <li>$y_{i}$ - фактическое значение зависимой переменной для $i$-го наблюдения</li>
        <li>$\hat{y}_i$ - предсказанное значение зависимой переменной для $i$-го наблюдения</li>
        <li>а именно возводя разности в квадрат, мы учитываем их абсолютные величины и их направления, чтобы избежать компенсации положительных и отрицательных отклонений друг от друга. Этот шаг и минимизирует сумму квадратов ошибок отклонений</li>
    </ul>
    <li><b>Решение уравнений:</b> Для нахождения оптимальных значений коэффициентов используются методы оптимизации, такие как градиентный спуск или аналитическое решение, которое можно найти путем дифференциации функции потерь по параметрам модели и приравнивании производных к нулю.</li>
    <li><b>Оценка модели:</b> После нахождения оптимальных коэффициентов модель оценивается на тестовых данных, и её производительность проверяется с использованием различных метрик, таких как коэффициент детерминации $R^2$ или средняя квадратическая ошибка.</li> <br>
<i>Метод наименьших квадратов широко используется из-за своей простоты и эффективности в оценке параметров линейной регрессии.</i>
</ol>

<h3>Математическое обоснование МНК</h3>

Для суммы квадратов ошибок отклонения существует математическое обоснование: <br>
Часто результатами реальных измерений имеет стандратное гаусское отклонение относительно измеряемого параметра:

В линейной регрессии сумма квадратов остатков (SSE) используется как мера ошибки модели, и ее минимизация имеет математическое обоснование, основанное на предположениях о распределении ошибок. Одно из основных предположений линейной регрессии - это предположение о нормальном распределении ошибок (остатков). Это означает, что остатки модели должны иметь стандартное гауссово (нормальное) распределение, то есть нормальное распределение с нулевым средним и постоянной дисперсией, где:
<ul>
<li>Нулевое среднее - означает, что среднее значение всех ошибок равно нулю. Это предположение означает, что модель в среднем правильно предсказывает зависимую переменную.</li>
<li>Постоянная дисперсия - означает, что дисперсия (разброс) ошибок остается постоянной независимо от значений независимых переменных. Это позволяет получить эффективные и консистентные оценки параметров модели при использовании МНК.</li>
</ul>
    Когда остатки имеют нормальное распределение, минимизация суммы квадратов остатков (SSE) приводит к нахождению наилучших оценок параметров модели. Это математически обосновано, потому что метод наименьших квадратов (МНК), который используется для оценки параметров, дает оценки параметров, которые являются наилучшими линейными несмещенными оценками при нормальном распределении ошибок. Таким образом, предположение о нормальном распределении ошибок обеспечивает математическое обоснование использования суммы квадратов остатков в качестве меры ошибки модели в линейной регрессии.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Задаем диапазон значений для оси x
x = np.linspace(-4, 4, 1000)

# Вычисляем плотность вероятности для стандартного нормального распределения
y = norm.pdf(x, 0, 1)  # Среднее значение = 0, стандартное отклонение = 1

# Строим график плотности вероятности
plt.figure(figsize=(8, 6))
plt.plot(x, y, 'b-', linewidth=2)
plt.title('Стандартное нормальное распределение')
plt.xlabel('Значение')
plt.ylabel('Плотность вероятности')
plt.xlabel('$f(x_i)$')
plt.ylabel('0')

# Добавляем горизонтальные и вертикальные оси
plt.axhline(0, color='black', linewidth=1.5)
plt.axvline(-4, color='black', linewidth=1.5)
plt.axvline(x=0, color='r', linestyle='-', linewidth=1)

# Создаем случайные значения для random_x
random_x = np.random.normal(0, 1, 6)  # Пример: случайные значения из стандартного нормального распределения

# Рисуем точки
plt.scatter(random_x, np.zeros_like(random_x), color='red', marker='o')

plt.grid(True)
plt.show()
```

Из выше сказанного и изображенного графика нормального распределения ошибок можно сделать вывод, что:</br>
<ul>
<li>Предположение о нормальном распределении ошибок означает, что чем выше точка $y_{max}$ значении $f(x_i)$, тем выше вероятность того, что измерения попадут ближе к области измеряемого параметра $f(x_i)$.</li>
<li>Из этого следует, что  вероятность того, что измерения попадут ближе к области измеряемого параметра $f(x_i)$ выше, чем у, тех у кого большее отклонение от $f(x_i)$, потому что нормальное распределение имеет колоколообразную форму, и наиболее вероятные (наиболее плотные) значения концентрируются вокруг среднего значения.</li>
</ul>
Нормальное распределение с нулевым средним и постоянной дисперсией - обеcпечивают к нахождению наилучших оценок параметров модели. Это позволяет минимизировать сумму квадратов остатков (SSE) и находить параметры модели, которые лучше всего соответствуют данным и имеют наибольшую вероятность быть "истинными" параметрами в генеральной совокупности.


стандартное гауссово (нормальное) распределение описывает формула: $$\omega(y_{i} | f(x_{i})) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(y_{i} - f(x_{i}))^2}{2\sigma^2}}$$

Где:
<ul>
    <li>$\sigma$ - стандартное отклонение. Это мера разброса значений случайной величины, чем меньше это значение, тем точнее прогнозы модели. влияет на ширину колокообразной формы графика. </li>
    <li>$f(x_{i})$ - математическое ожидание. Это значение определяет максимальную точку графика. Относительного этого параметра определяется разброс реальных измерений(красные точки) </li>
    <li>$\omega(y_{i} | f(x_{i}))$ - обозначение стандартного гаусского отклонения для значения $y_{i}$, при условии $f(x_{i})$. </li>
</ul>

<h3>Плотность распределения вероятности</h3>
<p>Плотность распределения вероятности (PDF) - это функция, которая описывает вероятность случайной величины принять определенное значение. Другими словами, она определяет, как вероятность распределена по всем возможным значениям случайной величины.<br>
Формально, для непрерывной случайной величины $X$, плотность распределения вероятности $f(x)$ определяется так, что вероятность того, $X$ попадет в некоторый интервал $[a,b]$, равна интегралу от $f(x)$ по интервалу: $$P(a \leq X \leq b) = \int_{a}^{b} f(x) \, dx $$
    Графически, плотность распределения вероятности часто представляется кривой на графике, где площадь под кривой в пределах интервала отражает вероятность попадания случайной величины в этот интервал. Плотность распределения вероятности может быть различной в зависимости от типа распределения (например, нормальное, равномерное, экспоненциальное и т. д.).<br>
    Плотность распределения вероятности часто используется для моделирования ошибок или остатков модели, чтобы оценить вероятностные характеристики этих ошибок и провести статистические выводы о параметрах модели.<br>
    <b>Например: </b><br>
Давайте вычислим вероятность $P(a \leq X \leq b)$ для стандартного нормального распределения, используя интегрирование. Мы можем сделать это, интегрируя функцию плотности вероятности $f(x)$ на интервале [1;-1]:

```python
import numpy as np
from scipy.integrate import quad

# Функция плотности вероятности стандартного нормального распределения
def pdf(x):
    return (1 / np.sqrt(2 * np.pi)) * np.exp(-0.5 * x**2)

# Определяем пределы интегрирования
a = -1
b = 1

# Вычисляем вероятность с помощью интегрирования
probability, _ = quad(pdf, a, b)

print("Вероятность P(-1 <= X <= 1):", probability)
```

Этот код использует функцию quad из библиотеки scipy.integrate, чтобы проинтегрировать функцию плотности вероятности pdf(x) на интервале от -1 до 1. Результат интегрирования дает вероятность того, что случайная величина попадает в этот интервал.
Такое условие может возникнуть, например, при анализе результатов экспериментов или измерений, когда требуется определить вероятность того, что случайная величина находится в определенном диапазоне значений. На практике это может быть полезно для оценки вероятности того, что наблюдаемые данные находятся в пределах определенного диапазона, что может иметь значение, например, при принятии решений на основе результатов исследований или анализе рисков.

<h3>Доказательство стандартного гауссово (нормального) распределения</h3>
Стандартное нормальное распределение (или стандартное гауссовское распределение) является одним из примеров функции плотности распределения вероятности (PDF).
Для стандартного нормального распределения PDF имеет вид:$$f(x) = \frac{1}{\sqrt{2\pi}} e^{-\frac{x^2}{2}}
$$
где $x$ - значение случайной величины, а $\frac{1}{\sqrt{2\pi}} e^{-\frac{x^2}{2}}$ -  это функция, которая определяет вероятность того, что случайная величина равна $x$ или попадает в определенный диапазон значений.</br>
Таким образом, интеграл от $f(x)$ по интервалу от $a$ до $b$ $P(a \leq X \leq b) = \int_{a}^{b} f(x) \, dx $ представляет собой вероятность того, что случайная величина $X$ попадает в этот интервал, а значением этой вероятности является площадь под кривой плотности вероятности от $a$ до $b$.<br>
Если мы подставим $f(x)$ в функцию плотности вероятности стандартного гауссовского распределения $$f(x) = \frac{1}{\sqrt{2\pi}} e^{-\frac{x^2}{2}}$$ вместо переменной $x$, мы получим: $$f(f(x_i)) = \frac{1}{\sqrt{2\pi}} e^{-\frac{f(x_i)^2}{2}}$$
где $f(x_{i})$ - это значение функции $f(x)$ для конкретного значения $x_{i}$.<br>
Тогда, если мы рассмотрим $y_{i}$ как случайную величину, имеющую нормальное распределение с параметрами $f(x_{i})$ и стандартным отклонением $\sigma$, мы можем записать плотность вероятности для $y_{i}$ как: 

$$\omega(y_{i} | f(x_{i})) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(y_{i} - f(x_{i}))^2}{2\sigma^2}}$$
</p>
Это означает, что $y_{i}$ имеет нормальное распределение с математическим ожиданием $f(x_{i})$ и стандартным отклонением $\sigma$.

<h3>Функция совместной плотности вероятности (для двух случайных величин)</h3>
При перемножении плотности вероятности $\omega(y_{i} | f(x_{i}))$ саму на себя, мы получаем функцию, которая описывает совместное распределение двух случайных величин $y_{i}$ и $f(x_{i})$:
$$ \prod_{i=1}^N \omega(y_i, f(x_i)) = \prod_{i=1}^N \sigma{\sqrt{2\pi}} \cdot \exp\left(-\frac{1}{2\sigma^2} \sum_{i=1}^N (y_i - f(x_i))^2\right)
$$
Где:
<ul>
    <li>$\prod_{i=1}^N \sigma{\sqrt{2\pi}}$ - множитель функции являющийся константой, которая умножает функцию. Он отвечает за масштабирование и нормализацию, чтобы плотность вероятности была правильно отмасштабирована и интегрировалась до единицы. </li>
    <li>$\exp$ - экспонента, описывающую степень, в которую возводится основание экспоненты $e$. В данном случае, экспонента содержит сумму квадратов отклонений $(y_i - f(x_i))^2$.</li>
    <li>$(-\frac{1}{2\sigma^2} \sum_{i=1}^N (y_i - f(x_i))^2)$ - степень экспоненты, которая устанавливает масштаб и форму экспоненциальной функции, влияя на форму и крутизну пика вероятностного распределения.</li>
</ul>
Условие предпологает что мы интересуемся тем, какие значения могут одновременно принимать обе этих переменные и с какой вероятностью. Сама формула  учитывает ошибки предсказания модели, вычисляя квадраты разностей между фактическими значениями $y_{i}$ и предсказанными значениями $f(x_{i})$. Это позволяет нам оценить, насколько хорошо модель соответствует данным и какие значения случайных переменных могут быть наиболее вероятными в нашем наборе данных. Это важный шаг в статистике и машинном обучении, особенно в контексте оценки параметров моделей, таких как линейная регрессия.

Очевидно, что при поиске максимума можно не учитывать множитель, так как он не зависит от параметров модели и не влияет на положение максимума.$$\prod_{i=1}^N \sigma{\sqrt{2\pi}}$$ А экспонента будет достигать максимальной минимазиции квадратов при минимуме ее степени $(-\frac{1}{2\sigma^2} \sum_{i=1}^N (y_i - f(x_i))^2)$,  т.к. перед степенью экспоненты стоит знак минус.<br>
Наша задача состоит в том, чтобы подобрать $f(x)$ так, чтобы выражение $-\frac{1}{2\sigma^2} \sum_{i=1}^{N} (y_i - f(x_i))^2 \rightarrow \min$ стремилось к минимуму. И получается, что лучшее описание экспериментальных данных с помощью функции $f(x)$  должно проходить по точкам, в которых достигается минимум этого выражения. Для этого мы можем отбросить положительный множитель $-\frac{1}{2\sigma^2}$ сказать, что нам нужен минимум квадрат отклонений $E = \sum_{i=1}^{N} (y_i - f(x_i))^2 \rightarrow \min$



<h2>Решение задач аппроксимации</h2>
<b>Аппроксимация</b> - это процесс нахождения функции или модели, которая приближенно описывает набор данных или некоторую функцию. Целью аппроксимации является создание простой модели или функции, которая может использоваться для прогнозирования или объяснения данных. В контексте линейной регрессии аппроксимация заключается в нахождении линейной функции (или другой модели), которая наилучшим образом соответствует наблюдаемым данным.</br>
Минимизация SSE означает, что аппроксимация лучше соответствует наблюдаемым данным.<br>
<i><b>Цель задачи</b></i> подобрать свободные параметры некой функции $f(x)$ с параметрами $a,b,c,...$ : $$f(x;a,b,c,..)$$
таким образом чтобы критерий качества принимал минимальные значения, то есть, величина E зависит от этих подбираемых величин:
$$E(a,b,c,...) = \sum_{i=1}^{N} (y_i - f(x_i;a,b,c))^2$$
Где $E$ - и есть критерий качества минимум суммы квадрата ошибок(SSE) c параметрами $a,b,c,...$ <br>
По сути, критерий качества $E$ для задачи аппроксимации можно представить график квадратичной функции от аргументов $a,b,c,...$:

```python
import numpy as np
import matplotlib.pyplot as plt

# определение функции E(x)
def E(x):
    return x**2

# определение производной dE/dx
def dE_dx(x):
    return 2*x

# создание массива значений x
x = np.linspace(-10, 10, 400)

# вычисление значений E(x) и dE/dx
y = E(x)
dy_dx = dE_dx(x)

# нахождение минимума E(x)
x_star = x[np.argmin(y)]

# создание графика
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(x, y, 'r', label='E(x)')
ax.plot(x_star, E(x_star), 'bo')
ax.axhline(y=E(x_star), color='g', linestyle='--')
ax.set_xlabel('X')
ax.set_ylabel('E(X)')
ax.set_title('График функции E(x) и ее минимума')
ax.legend()
ax.grid(True)
plt.show()
```

Точка экстремума(минимума) данной функции и есть точка, в которой производная равна нулю. Эта точка и будет соответствовать оптимальным параметром $a,b,c,...$ для критерий качества $E$. Для ее нахождения достаточно решить уравнение: $$\frac{dE(x)}{dx} = 0$$


Для продожения решения задчи необходимо взять частные производные по каждому параметру $a,b,c,...$ и приравнять результат нулю, получим систему линейных уравнений: 
$$\begin{cases}
\frac{\partial E(a,b,c,...)}{\partial a} = \sum_{i=1}^{N} \left( y_i - f(x_i;a,b,c,...) \right) \frac{\partial f(x_i)}{\partial a} = 0, \\
\frac{\partial E(a,b,c,...)}{\partial b} = \sum_{i=1}^{N} \left( y_i - f(x_i;a,b,c,...) \right)  \frac{\partial f(x_i)}{\partial b} = 0, \\ ...
\end{cases}$$
Где:
<ul>
    <li><b>Частная производная по параметру</b> $a$ :<br>
    $\frac{\partial E(a,b,c,...)}{\partial a} = \sum_{i=1}^{N} \left( y_i - f(x_i;a,b,c,...) \right) \frac{\partial f(x_i)}{\partial a} = 0$ <br>
    Это уравнение означает, что мы ищем такое значение параметра $a$, при котором изменение функции ошибок $E$ по $a$, при котром изменение функции ошибки $E$ по $a$ равно нулю. Для этого мы вычисляем разность между реальным значением $y_{i}$ и предсказанным значением $f(x_{i};a,b,c...)$ для каждого $i$-го наблюдения, умножаем умножаем разность на производную функции $f$ по каждому $a$ при $x_{i}$ и суммируем по всем $N$ наблюдениям. Полученная сумма должна быть равна нулю.
    </li>
    <li><b>Частная производная по параметру b:</b><br>
    $\frac{\partial E(a,b,c,...)}{\partial b} = \sum_{i=1}^{N} \left( y_i - f(x_i;a,b,c,...) \right)  \frac{\partial f(x_i)}{\partial b} = 0$<br>
    Это аналогичное уравнение для параметра $b$, при котром мы также ищем такое значение параметра $b$, при котром изменение функции ошибки $E$ по $b$ равно нулю. Для этого мы вычисляем разность между реальными значением $y_{i}$ и предсказанным значением $f(x_{i};a,b,c...)$ для каждого $i$-го наблюдателя, мы умножаем эту разность на производную функции $f$ по $b$ при $x_{i}$ и суммируем по всем $N$ наблюдениям. Полученная сумма также должна быть равна нулю.
    </li>
    
</ul>
Для понимания, необходимо наполнить конкретикой эту систему, нам нужно вернуться к исходному примеру с линейной функцией: $f(k,x,b)=kx+b$
Эта функция зависит от двух парпаметров: $k$ и $b$. В функции $f(x;k,b)$, $x$ представляет собой независимую переменную, то есть переменную, от которой зависит функция $f$, а $k$ и $b$ представляют собой параметры или коэффиценты, котоыре определяют форму или характер функции $f$. И частная производная к каждому из этих параметров будут следующими: 
$$\frac{\partial f(x_i)}{\partial k} = x_i; \quad \frac{\partial f(x_i)}{\partial b} = 1$$
Частная производная $f(x;k,b)$ по параметру $k$, обозначаемая как $\frac{\partial f}{\partial k}$, определяется как изменение функции $f$ по $k$, при условии, что все остальные переменные остаются постоянными. Аналогично, частная производная по параметру $b$, обозначаемая как $\frac{\partial f}{\partial b}$, показывает изменение функции $f$ по $b$, при постоянном значении остальных переменных. Параметр $k$ обычно называются наклоном(slope), а параметр $b$ - сдвигом(intercept).

Для функции $f(x;k,b)=kx+b$ вычисление частной производной по переменной $b$ означает, что мы рассматриваем переменную kx как константу. Так как при дифференцировании по константе получаем ноль, то частная производная по $b$ равна 1, поскольку переменная $b$ просто суммируется с $kx$.

Частная производная функции $f(x;k,b)=kx+b$ по переменной $k$ будет равна $x$. При вычислении этой производной по $k$ переменная $b$ рассматривается как константа, а коэффицент $k$ при $x$ отбрасывается по правилу дифферинцирирования.


Эти производные позволят нам найти оптимальные значения параметров $k$ и $b$ для минимизации критерия качества $E$.

Далее значение производных мы подставляем в систему:<br>
$$\begin{cases}
\frac{\partial E(a,b,c,...)}{\partial a} = \sum_{i=1}^{N} \left( y_i - f(x_i;a,b,c,...) \right) \frac{\partial f(x_i)}{\partial a} = 0, \\
\frac{\partial E(a,b,c,...)}{\partial b} = \sum_{i=1}^{N} \left( y_i - f(x_i;a,b,c,...) \right)  \frac{\partial f(x_i)}{\partial b} = 0, \\ ...
\end{cases}$$
мы получаем: $$\left\{
\begin{array}{ll}
\sum_{i=1}^{N} (- kx_{i} - b) \cdot x_{i} = 0 \\
\sum_{i=1}^{N} (- kx_{i} - b) \cdot 1 = 0
\end{array}
\right.$$
Где:
<ul>
    <li>$N$ - количество эксперментов или же число наблюдений $y_{i}$</li>
    <li>$-kx_{i} - b$ - значение подставленной функции $f(x_{i})$</li>
    <li>$x_{i}$ - частная производная для параметра $k$</li>
    <li>$1$ - частная производная для параметра $b$</li>
    
</ul>

Далее, предудущую систему мы можем записать в виде:
$$\begin{cases}
\sum_{i=1}^{N} y_i \cdot x_i - k \cdot \sum_{i=1}^{N} x_i^2 - b \cdot \sum_{i=1}^{N} x_i = 0 \\
\sum_{i=1}^{N} y_i - k \cdot \sum_{i=1}^{N} x_i - b \cdot N = 0
\end{cases}
$$
Здесь мы отдельно расписали сумму для $y_{i}$, $k$ и $b$ и умножили каждую на $x_{i}$

Для решения необходимо разделить все на $N$:
$$\begin{align*}
\left\{
\begin{array}{ll}
\frac{1}{N}\sum_{i=1}^{N}y_i \cdot x_i - k \cdot \frac{1}{N}\sum_{i=1}^{N}x_i^2 - b \cdot \sum_{i=1}^{N}x_i^2 - b \cdot \frac{1}{N} \sum_{i=1}^{N}x_i = 0 \\
\\
\frac{1}{N}\sum_{i=1}^{N}y_i - k \cdot \frac{1}{N}\sum_{i=1}^{N}x_i - b= 0
\end{array}
\right.
\end{align*}
$$
Данные выражения, которые записаны под суммой характерны для теории вероятности. Формулы с суммами представляют собой первые и вторые начальные моменты, а также один смешанный момент:
